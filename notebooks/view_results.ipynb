```python
path = os.path.join(os.path.dirname(os.getcwd()), 'results')
```

```python
plt.rcParams['figure.figsize'] = (20, 10)
```

```python
def plotting(path, title, target):
    for currdir, subdirs, files in os.walk(path):
        if title in currdir:
            for filename in files:
                if target in filename:
                    data = np.load(os.path.join(currdir, filename))
                    plt.plot(data, label=filename.split('_')[3:])
    plt.title(f'{title}: {target}')
    plt.legend()
    plt.show()
```
