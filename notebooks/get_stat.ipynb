```python
import os
import numpy as np
```

```python
path = os.path.join(os.path.dirname(os.getcwd()), 'tmp')
```

```python
target = 'te_acc'
cnt = 0

filelist = sorted(os.listdir(path))

for f in filelist:
    if f.endswith(f'{target}.npy'):
        data = np.load(os.path.join(path, f))
        cnt += 1
        print(f'[{cnt:>2d}] {f}\n  |-- Acc: Max {np.max(data) * 100:.4f}% ({np.argmax(data)} round) | Last {data[-1] * 100:.4f}% | Avg {np.mean(data) * 100:.4f}% ({np.argmax(data > np.mean(data))} round)\n')
```
